```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("play_tennis.csv")
data
```

```python
model = DecisionTreeClassifier()
```

```python
data.outlook = data.outlook.map({'Sunny':0,"Overcast":1,"Rain":2})
data.temp = data.temp.map({'Hot':0,'Mild':1,'Cool':2})
data.humidity = data.humidity.map({'High':0,"Normal":1})
data.wind = data.wind.map({'Strong':0,'Weak':1})
data.play = data.play.map({'Yes':1,'No':0})
```

```python
data.head()
```

```python
X=data.drop(columns=["day","play"])
```

```python
y=data["play"]
```

```python
X,y
```

```python
model.fit(X,y)
```

```python
y_pred=model.predict(X)
```

```python
model.score(X,y)
```

```python
np.mean(y==y_pred)
```
